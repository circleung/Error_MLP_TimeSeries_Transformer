```python
import pandas as pd
import numpy as np

def train_test_split_by_scenario(df: pd.DataFrame, split: float = 0.9, shuffle: bool = False, random_state: int = None):
    """
    Split the dataframe into train and test sets by scenario_number.
    Entire scenarios go either into train or test.

    Parameters
    ----------
    df : pd.DataFrame
        The dataframe with a 'scenario_number' column.
    split : float, optional (default=0.9)
        Fraction of scenarios to include in the train set.
    shuffle : bool, optional (default=False)
        Whether to shuffle the scenarios before splitting.
    random_state : int, optional (default=None)
        Random seed for reproducibility when shuffle=True.

    Returns
    -------
    train_df : pd.DataFrame
        Data for training (split% of scenarios).
    test_df : pd.DataFrame
        Data for testing (remaining scenarios).
    """
    scenarios = df["scenario_number"].unique()

    if shuffle:
        rng = np.random.default_rng(seed=random_state)
        rng.shuffle(scenarios)

    n_train = int(len(scenarios) * split)

    train_scenarios = scenarios[:n_train]
    test_scenarios = scenarios[n_train:]

    train_df = df[df["scenario_number"].isin(train_scenarios)]
    test_df = df[df["scenario_number"].isin(test_scenarios)]

    return train_df, test_df
```

```python
train_df, test_df = train_test_split_by_scenario(df, split=0.9, shuffle = True, random_state = 42)
```
